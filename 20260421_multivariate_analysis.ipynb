```python
import pandas as pd
from sympy import symbols, Integral
```

# Discrete Random Variable

```python
df_discrete = pd.DataFrame({
    'x': [1, 2, 3, 4, 5, 6],           # 出目
    'p': [1/6, 1/6, 1/6, 1/6, 1/6, 1/6] # 各確率 p(x)
})

# E(X) = Σ (x * p(x))
expected_value = (df_discrete['x'] * df_discrete['p']).sum()

print(f"Expected Variable: {expected_value:.1f}")
```

# Continuous Random Variable

```python
x = symbols('x')
f = 2 * x  # 確率密度関数

# ∫ x * f(x) dx (範囲 0 to 1)
expectation_expr = Integral(x * f, (x, 0, 1))

print(f"Expected Variable: {expectation_expr.doit()}")
```

# Example 2.13

```python
import numpy as np

# 1. 同時確率関数 p12(x1, x2) の定義 (行: x1={-1, 0, 1}, 列: x2={0, 1})
P = np.array([
    [0.24, 0.06],
    [0.16, 0.14],
    [0.40, 0.00]
])

x1 = np.array([-1, 0, 1])
x2 = np.array([0, 1])

# 2. 周辺確率分布の計算 (マージンをとる)
p1 = P.sum(axis=1) # 行方向の合計 -> p1(x1)
p2 = P.sum(axis=0) # 列方向の合計 -> p2(x2)

# 3. 平均 μ1, μ2 の計算
mu1 = np.sum(x1 * p1)
mu2 = np.sum(x2 * p2)

# 4. 分散 σ11, σ22 の計算
sig11 = np.sum((x1 - mu1)**2 * p1)
sig22 = np.sum((x2 - mu2)**2 * p2)

# 5. 共分散 σ12 の計算
# 全ペア (x1, x2) に対して (x1-μ1)*(x2-μ2)*p(x1,x2) を計算
X1, X2 = np.meshgrid(x1, x2, indexing='ij')
sig12 = np.sum((X1 - mu1) * (X2 - mu2) * P)

# 6. 分散共分散行列 Σ の構築
Sigma = np.array([
    [sig11, sig12],
    [sig12, sig22]
])

print(f"Mean Vector μ: [{mu1:.2f}, {mu2:.2f}]")
print("Covariance Matrix Σ:")
print(Sigma)
```
