```python
import pandas as pd
from sympy import symbols, Integral
```

# Discrete Random Variable

```python
df_discrete = pd.DataFrame({
    'x': [1, 2, 3, 4, 5, 6],           # 出目
    'p': [1/6, 1/6, 1/6, 1/6, 1/6, 1/6] # 各確率 p(x)
})

# E(X) = Σ (x * p(x))
expected_value = (df_discrete['x'] * df_discrete['p']).sum()

print(f"Expected Variable: {expected_value:.1f}")
```

# Continuous Random Variable

```python
x = symbols('x')
f = 2 * x  # 確率密度関数

# ∫ x * f(x) dx (範囲 0 to 1)
expectation_expr = Integral(x * f, (x, 0, 1))

print(f"Expected Variable: {expectation_expr.doit()}")
```
